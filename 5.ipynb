# Import libraries

```python
import numpy as np
from sklearn.datasets import load_breast_cancer, load_diabetes
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor
from sklearn.metrics import accuracy_score, mean_squared_error, r2_score
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
```

# Classification

## Load dataset

```python
data = load_breast_cancer()
X = data.data
y = data.target
```

## Train-Test-Split

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)
```

## Decision Tree Classifier

```python
dt_clf = DecisionTreeClassifier(random_state=42)
dt_clf.fit(X_train, y_train)
```

## Performance

```python
print("Decision Tree Accuracy (Train):",
      accuracy_score(y_train, dt_clf.predict(X_train)))
print("Decision Tree Accuracy (Test):",
      accuracy_score(y_test, dt_clf.predict(X_test)))
```

## Cost Complexity Pruning

```python
path = dt_clf.cost_complexity_pruning_path(X_train, y_train)
ccp_alphas = path.ccp_alphas

pruned_clf = DecisionTreeClassifier(ccp_alpha=ccp_alphas[5], random_state=42)
pruned_clf.fit(X_train, y_train)

print("\nAfter Pruning:")
print("Pruned Tree Accuracy (Train):",
      accuracy_score(y_train, pruned_clf.predict(X_train)))
print("Pruned Tree Accuracy (Test):",
      accuracy_score(y_test, pruned_clf.predict(X_test)))
```

## Random Forest

```python
rf_clf = RandomForestClassifier(n_estimators=100, random_state=42)
rf_clf.fit(X_train, y_train)

print("\nRandom Forest Accuracy (Train):",
      accuracy_score(y_train, rf_clf.predict(X_train)))
print("Random Forest Accuracy (Test):",
      accuracy_score(y_test, rf_clf.predict(X_test)))
```

## Adaboost with Decision Stumps

```python
stump = DecisionTreeClassifier(max_depth=1, random_state=42)

ada_clf = AdaBoostClassifier(
    estimator=stump,
    n_estimators=100,
    random_state=42
)

ada_clf.fit(X_train, y_train)

print("\nAdaBoost Accuracy (Train):",
      accuracy_score(y_train, ada_clf.predict(X_train)))
print("AdaBoost Accuracy (Test):",
      accuracy_score(y_test, ada_clf.predict(X_test)))
```

# Regression

## Load Dataset

```python
diabetes = load_diabetes()
X_reg = diabetes.data
y_reg = diabetes.target
```

## Train-Test Split

```python
X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(
    X_reg, y_reg, test_size=0.3, random_state=42
)
```

## Decision Tree Regressor

```python
dt_reg = DecisionTreeRegressor(random_state=42)
dt_reg.fit(X_train_reg, y_train_reg)
```

## Performance

```python
print("Decision Tree R2 (Train):",
      r2_score(y_train_reg, dt_reg.predict(X_train_reg)))
print("Decision Tree R2 (Test):",
      r2_score(y_test_reg, dt_reg.predict(X_test_reg)))

print("Decision Tree RMSE (Test):",
      np.sqrt(mean_squared_error(y_test_reg,
                                 dt_reg.predict(X_test_reg))))
```

## Cost Complexity Pruning

```python
path_reg = dt_reg.cost_complexity_pruning_path(X_train_reg, y_train_reg)
ccp_alphas_reg = path_reg.ccp_alphas

pruned_reg = DecisionTreeRegressor(
    ccp_alpha=ccp_alphas_reg[10],
    random_state=42
)
pruned_reg.fit(X_train_reg, y_train_reg)

print("\nAfter Pruning:")
print("Pruned Tree R2 (Train):",
      r2_score(y_train_reg, pruned_reg.predict(X_train_reg)))
print("Pruned Tree R2 (Test):",
      r2_score(y_test_reg, pruned_reg.predict(X_test_reg)))
```

## Random Forest Regressor

```python
rf_reg = RandomForestRegressor(n_estimators=100, random_state=42)
rf_reg.fit(X_train_reg, y_train_reg)

print("\nRandom Forest R2 (Train):",
      r2_score(y_train_reg, rf_reg.predict(X_train_reg)))
print("Random Forest R2 (Test):",
      r2_score(y_test_reg, rf_reg.predict(X_test_reg)))

print("Random Forest RMSE (Test):",
      np.sqrt(mean_squared_error(y_test_reg,
                                 rf_reg.predict(X_test_reg))))
```
